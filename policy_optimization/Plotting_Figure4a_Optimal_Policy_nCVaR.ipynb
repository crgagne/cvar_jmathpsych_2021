## Setup Agent

```python
p,maze = setup_params(4) 
agent = Agent({'num_states':Ns-1,'num_actions':Na,'start_states':[4],'r_params':p['r_params']},maze)
```

# Further Processing

## Policy in Correct Format

```python
# convert policy to right format 
Pol = np.zeros((Ns,Ny,3))
for x in range(Ns):
    for y in range(Ny):
        Pol[x,y,Pol_mat[x,y]-1]=1
        
Pol.shape
# print(Pol[0:9,4,:])
# print(Pol[0:9,3,:])
# print(Pol[0:9,2,:])
# print(Pol[0:9,1,:])
# print(Pol[:,1,:])
# print(Pol[:,0,:])
```

## Policy Induced Transition Matrix P(x'|x,y)

```python
# creating transition probability matrix induced by optimal policy 
P_ssy= induce_P(P,Pol,Ny,Ns,
                 None,fixed_y=False,
                 pol_type='stochastic',Na=3) # XXXX This line changes to Pol
print(P_ssy.shape)
np.round(P_ssy[:,:,2],2)
```

## Convert some stuff for Plotting

```python
V_py = convert_SI_function(V,agent,neg_term_states=[],num_inter_points=Ny)
V_py.shape
```

## Effective Policy

```python
Pol_effective = np.zeros_like(Pol) # this is going to be where you end up in X probabilistically 
for s in range(Ns-1):
    for y in range(Ny):
        if s==0:
            Pol_effective[s,y,:] = [0,0,1]
        elif s==Ns-2:
            Pol_effective[s,y,:] = [P_ssy[s,s,y],P_ssy[s,s,y],P_ssy[s,s-1,y]] # can't go to the right 
        else:
            Pol_effective[s,y,:] = [P_ssy[s,s,y],P_ssy[s,s+1,y],P_ssy[s,s-1,y]]
            
# two quick fixes to code above, because the right most state duplicates for Ns-2
Pol_effective[8,1,:]=[0.06, 0.90, 0.04] # fix duplicate for staying v going right 
Pol_effective[8,2,:]=[1/3,1/3,1/3] # fix duplicate for staying v going right 
```
