## Load MDP from Matlab

```python
# load old matlab results, which have useful info for MDP
mat = loadmat('../simulation_results/2D_optimal_policy.mat')

# Unpack 
Ny = mat['Y_set_all'].shape[1]
Ns = mat['Y_set_all'].shape[0] # could subtract one here 
Na = 4
gamma = mat['dis'][0][0]
y_set = mat['Y_set_all'][0,:]
P = mat['P']
r = mat['r']
print(P.shape)
print("x,x',a")
y_set
```

```python
testcase = 1
p,maze = setup_params(testcase) 
maze[4,0]=1
maze[4,6]=1

agent = Agent({'num_states':Ns-1,
               'num_actions':Na,
               'start_states':[4],'r_params':p['r_params']},maze)
```

```python
# Load 
container = np.load(fname)
data = [container[key] for key in container]
Vd = data[0]
Xid = data[1]
V_storaged=data[2]
Qd = data[3]
Q_storaged = data[4]
Xid = data[5]
Xi_storaged = data[6]
```

## Load pCVaR

```python
fname = '../simulation_results/BernCase_1S_optimal_nonprioritized_VI.npz'
fname = '../simulation_results/2D_optimal_policy_pCVaR.npz'
fname
```

```python
# Load 
container = np.load(fname)
data = [container[key] for key in container]
V = data[0]
Xi = data[1]
V_storage=data[2]
Q = data[3]
Q_storage = data[4]
Xi = data[5]
Xi_storage = data[6]
```

## Optimal Actions

```python
roundoff = 5
Pol_d = np.zeros_like(Qd)
Pol = np.zeros_like(Qd)
for yi in range(len(y_set)):
    for x in range(Ns):
        
        
        # round to roundoff decimmals
        minn = np.min(np.round(Q[x,yi,:],roundoff))
        Pol[x,yi,np.round(Q[x,yi,:],roundoff)==minn]=1
    
        minn = np.min(np.round(Qd[x,yi,:],roundoff))
        Pol_d[x,yi,np.round(Qd[x,yi,:],roundoff)==minn]=1
        
        #Pol_d[x,yi,np.argmin(Qd[x,yi,:])]=1
```

```python
V.shape
```

```python
V_py = convert_SI_function(V,agent,neg_term_states=[],num_inter_points=6)
Vd_py = convert_SI_function(Vd,agent,neg_term_states=[],num_inter_points=6)

Pol_py = convert_SIA_function(Pol,agent,neg_term_states=[],num_inter_points=6)
Pol_d_py = convert_SIA_function(Pol_d,agent,neg_term_states=[],num_inter_points=6)
```
