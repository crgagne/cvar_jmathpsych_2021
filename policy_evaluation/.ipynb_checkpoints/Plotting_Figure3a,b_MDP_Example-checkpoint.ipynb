## Load MDP Information

```python
testcase=4
subcase='aS'
```

```python
fname = '../simulation_results/1D_random_policy_nCVaR.npz'
```

```python
# Load 
container = np.load(fname)
data = [container[key] for key in container]

V = data[0] # get values 
Xi = data[1] # get distortion weights 
V_storage = data[2]
```

```python
# Load CVaR Policy Eval on Random Code 
# It also has some stuff from the Value Iteration on the Same Case 
mat = loadmat('../simulation_results/1D_random_policy.mat')

# get MDP parameters 
Ny = mat['Y_set_all'].shape[1]
Ns = mat['Y_set_all'].shape[0] # could subtract one here 
Na = 3
gamma = mat['dis'][0][0]
y_set = mat['Y_set_all'][0,:]
P = mat['P']
r = mat['r']
r_fun = mat['r'][:,0]
gamma = mat['dis']
```

```python
# Random policy 
Pol= np.zeros((Ns,Ny,3))
Pol[:,:,:]=0.33
Pol[:,2,:]
```

## Plotting MDP

```python
p,maze = setup_params(testcase) 
agent = Agent({'num_states':Ns-1,'num_actions':Na,'start_states':[4],'r_params':p['r_params']},maze)
```

```python
plt.figure(figsize=(8,8/7),dpi=500)
ax = plt.gca()

# random policy
Pol_py = np.zeros((9,21,3))
Pol_py[:,:,1:3]=0.5

term_states_for_plotting=[0,9]

Qrange = [-10,10]

# Value Underlay 
plot_q_or_pi(np.zeros_like(Pol_py),np.zeros_like(maze),'',ax,agent.maze,q_or_pi='q',
            roundoff=1,inc_triangles=False,tri_add_labels=False,annot_value=False,Qrange=Qrange,
             term_states_for_plotting=term_states_for_plotting)

embellish_plot(ax,agent.maze,
    agent,None,
    cost=True,
    corner_labels=False,
    color_agent='b',
    center_rewards=False,
    r_fontsize=10,
    add_rewards=True,ec='white',fc='white',alpha=0.0,outer_lw=1,reward_color='k')

# Policy as Arrow 33$
plot_1D_arrows(Pol[:,0,:],
               0,
               maze,
               term_states_for_plotting,
               box=False,box_ec=(0.9,0.9,0.9,1),jointbox=True)

arr_img = mpimg.imread('../mscl/man_blue.png')
imagebox = OffsetImage(arr_img , zoom=.113)
ab = AnnotationBbox(imagebox, (4.0, 0),pad=0)
ax.add_artist(ab)


arr_img = mpimg.imread('../mscl/lava_pit.png')
imagebox = OffsetImage(arr_img , zoom=.013)
ab = AnnotationBbox(imagebox, (0, 0),pad=0)
ax.add_artist(ab)

fs = 14
plt.text(4,-0.7,r'$x_0$',fontsize=fs+2,zorder=3,va='center',ha='center',color='k')
```
