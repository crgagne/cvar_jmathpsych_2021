# Loading Data

## Loading Policy Evaluation

```python
# load policy eval results 
fname = '../simulation_results/1D_random_policy_nCVaR.npz'
container = np.load(fname)
data = [container[key] for key in container]

V = data[0] # get values 
Xi = data[1] # get distortion weights 
V_storage = data[2]
```

```python
# load old matlab results, which have useful info for MDP
mat = loadmat('../simulation_results/1D_random_policy.mat')

# get MDP parameters 
Ny = mat['Y_set_all'].shape[1]
Ns = mat['Y_set_all'].shape[0] # could subtract one here 
Na = 3
gamma = mat['dis'][0][0]
y_set = mat['Y_set_all'][0,:]
P = mat['P']
r = mat['r']
r_fun = mat['r'][:,0]
gamma = mat['dis']
```

```python
# Random policy 
Pol= np.zeros((Ns,Ny,3))
Pol[:,:,:]=1/3
Pol[:,2,:]
```

```python
Pol.shape
```

## Set-up an Agent

```python
testcase=4
subcase='aS'
p,maze = setup_params(testcase) 
agent = Agent({'num_states':Ns-1,'num_actions':Na,'start_states':[4],'r_params':p['r_params']},maze)
```

## Converting into Python Format (Same for 1D)

```python
# convert value function to Python format; won't matter for 1D
V_py = convert_SI_function(V,agent,neg_term_states=[],num_inter_points=Ny)
```

# Further processing

## Policy Induced Transition Matrix P(x'|x,y)

```python
# P(s|s,y)
P_ssy = induce_P(P,Pol,Ny,Ns,None,fixed_y=False,pol_type='stochastic',Na=Na) # XXXX This line changes to Pol
np.round(P_ssy[:,:,2],2)
```

```python
P_ssy.shape
```

```python
P_ssy[:,:,-1].sum(axis=1)
```

## Distorted Transition Matrix P(x', y'|x,y)*Xi(x'|x,y)  
- For a particular set of ys

```python
(P_xy_xpyp,P_xy_xpyp_alpha,Xi_xy_xpyp) = create_distorted_probability_matrix(P_ssy, #
                                        Xi, # same size as P_ssy
                                        Ns,Ny,Na,
                                        y_set,
                                        yis = [5,4,3,2,1,0],# which y values do you want for plotting
                                        adjust_y=False,
                                        )
```

# Plotting

```python
plot_1D_values_policy_distortions(Pol,
                                  V_py,
                                  agent,
                                  np.round(P_xy_xpyp_alpha,2), # round to 2nd
                                  np.round(Xi_xy_xpyp,2), # round to 2nd
                                  static_or_dynamic='static',
                                  y_set = [0,0.05,0.1,0.3,0.6,1],
                                  yis = [5,4,3,2,1,0],
                                  Qrange = [-10,10],
                                  term_states_for_plotting = [0,9],
                                  staybox='bottom',
                                  )
```

# Color Bar

```python
fig = plt.figure(figsize=(8,1),dpi=200)
ax = plt.gca()

# random policy
Pol_py = np.zeros((9,21,3))
Pol_py[:,:,1:3]=0.5

term_states_for_plotting=[0]

Qrange = [-10,10]

for t in term_states_for_plotting:
    Pol_py[t,:]=0
    
# Value Underlay 
v = -1*V[0:9,0].reshape(maze.shape,order='C')
_,im_value = plot_q_or_pi(np.zeros_like(Pol_py),v,'',ax,agent.maze,q_or_pi='q',
            roundoff=1,inc_triangles=False,tri_add_labels=False,annot_value=False,Qrange=Qrange,
             term_states_for_plotting=term_states_for_plotting,colorbar=True)
plt.gca().set_visible(False)
plt.colorbar(im_value,label=r'$nCVaR^{\pi}_{\bar{\alpha},x}$',orientation="horizontal")
```
