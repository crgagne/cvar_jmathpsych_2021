# Loading Data

```python
testcase=4
subcase='aS'
```

## Loading Policy Evaluation

```python
fname = '../simulation_results/1D_random_policy_pCVaR.npz'
```

```python
# Load 
container = np.load(fname)
data = [container[key] for key in container]

V = data[0] # get values 
Xi = data[1] # get distortion weights 
V_storage = data[2]
```

```python
# Load CVaR Policy Eval on Random Code 
# It also has some stuff from the Value Iteration on the Same Case 
mat = loadmat('../simulation_results/1D_random_policy.mat')

# get MDP parameters 
Ny = mat['Y_set_all'].shape[1]
Ns = mat['Y_set_all'].shape[0] # could subtract one here 
Na = 3
gamma = mat['dis'][0][0]
y_set = mat['Y_set_all'][0,:]
P = mat['P']
r = mat['r']
r_fun = mat['r'][:,0]
gamma = mat['dis']
```

```python
gamma
```

```python
# Random policy 
Pol= np.zeros((Ns,Ny,3))
Pol[:,:,:]=1/3
Pol[:,2,:]
```

```python
print(Pol.shape)
print(Xi.shape)
```

## Set-up an Agent

```python
p,maze = setup_params(testcase) 
agent = Agent({'num_states':Ns-1,'num_actions':Na,'start_states':[4],'r_params':p['r_params']},maze)
```

## Converting into Python Format (Same for 1D)

```python
# convert value function to Python format; won't matter for 1D
V_py = convert_SI_function(V,agent,neg_term_states=[],num_inter_points=Ny)
```

# Further processing

```python
# P(s|s,y)
P_ssy = induce_P(P,Pol,Ny,Ns,None,fixed_y=False,pol_type='stochastic',Na=Na) # XXXX This line changes to Pol
np.round(P_ssy[:,:,2],2)
```

# Color Bars and Arrows Figure 3c

```python
fig = plt.figure(figsize=(8,1),dpi=200)
ax = plt.gca()

# random policy
Pol_py = np.zeros((9,21,3))
Pol_py[:,:,1:3]=0.5

term_states_for_plotting=[0]

Qrange = [-10,10]

for t in term_states_for_plotting:
    Pol_py[t,:]=0
    
# Value Underlay 
v = -1*V[0:9,0].reshape(maze.shape,order='C')
_,im_value = plot_q_or_pi(np.zeros_like(Pol_py),v,'',ax,agent.maze,q_or_pi='q',
            roundoff=1,inc_triangles=False,tri_add_labels=False,annot_value=False,Qrange=Qrange,
             term_states_for_plotting=term_states_for_plotting,colorbar=True)
plt.gca().set_visible(False)
plt.colorbar(im_value,label='value \n'+r'$E^{\pi}_{x}$'+\
             ' or '+r'$ nCVaR^{\pi}_{\bar{\alpha},x}$'+\
             ' or '+r'$ pCVaR^{\pi}_{\alpha,x}$',orientation="horizontal")
```

```python
plt.figure(figsize=(4,1),dpi=500)

style = "Simple, tail_width=0.5, head_width=4, head_length=8"
rad = '0'
plt.ylim([0,1.5])

for i,p in enumerate([0.5,0.5]):
    
    if i==0:
        xstart = 0.3
        xoffset = -0.1
        ystart = 0.5
        yoffset = 1
        rad='0'
    elif i==1:
        xstart = 0.3
        xoffset = 0.15
        ystart = 0.5
        yoffset= 0.00
        rad = '-1'
        
    kw = dict(arrowstyle=style, color=(0.3,0.3,0.3,1)) #P_xy_xpyp_alpha[x,i,xps,yps]
    patch = patches.FancyArrowPatch(
                            (xstart,ystart), # start in center of square
                            (xstart+xoffset, # point towards difference 
                             ystart+yoffset), 
                             connectionstyle="arc3,rad="+rad,
                             linestyle='-', 
                                    linewidth=2.5,**kw,zorder=5)


    plt.gca().add_patch(patch)
    
    
plt.xlim([0,1])

sns.despine(left=True,bottom=True)
plt.xticks([])
plt.yticks([])
plt.text(0.35,-0.1,'distorted transitions\n'+r'$p^{\pi}(x^\prime|x)\xi(x^\prime|x)$'+'',ha='center',va='center',fontsize=14)

plt.show()

```

```python
plt.figure()

s_idcs = [0.5,0.5]

# plt.xlim([0.25,0.75])
# plt.ylim([0.1,0.75])
plt.ylim([0,1])
plt.xlim([0,1])

arrow_markers = np.array(['$\u21BB$','$\u2192$','$\u2190$'])
markers = arrow_markers # select all markers
mults = np.ones(len(markers))*0.33*1.2+.4 # make multiplier relative to random.
for marker,mult in zip(markers,mults):
    yoffset=-0.01
    xoffset=0
    if marker==arrow_markers[0]:
        #yoffset = 0.21 #-0.08
        #angle = 0
        yoffset = -0.2
        angle=0#195
        xoffset=-0.005
    elif marker==arrow_markers[1]:
        xoffset = 0.08 #0.14#+0.07
        angle=0
    elif marker==arrow_markers[2]:
        xoffset=-0.08 #0.14#-0.07
        angle=0

    t = mpl.markers.MarkerStyle(marker=marker)
    t._transform = t.get_transform().rotate_deg(angle)
    plt.scatter(s_idcs[1]+xoffset,s_idcs[0]+yoffset,
                marker=t,s=1000,color='k',zorder=10)#,ms=10*np.sqrt(mult))



# xcenter = s_idcs[1]
# ycenter = s_idcs[0]+0.1
# xleft = xcenter-0.28
# xright = xcenter+0.28
# ytop = ycenter-0.2
# ybottom = ycenter+0.2
xcenter = s_idcs[1]
ycenter = s_idcs[0]-0.1
xleft = xcenter-0.16 #0.28
xright = xcenter+0.16 #0.28
ytop = ycenter+0.2
ybottom = ycenter-0.2


xy = [[xleft,ytop],
      [xright,ytop], # top
      [xright,ycenter],
      [xcenter+(xright-xcenter)*2/5,ycenter], #
      [xcenter+(xright-xcenter)*2/5,ybottom], # right leg
      [xcenter+(xleft-xcenter)*2/5,ybottom],
      [xcenter+(xleft-xcenter)*2/5,ycenter],
      [xleft,ycenter]
     ]
xy =np.array(xy)
ax = plt.gca()
ax.add_patch(
patches.Polygon(
    xy=xy,  # point of origin.
    linewidth=1,
    facecolor='white',
    edgecolor='k',
    fill=True,
    zorder=2,
    closed=True,
)
)

plt.xticks([])
plt.yticks([])
sns.despine(bottom=True,left=True)
plt.text(0.5,0.0,'policy\n'+r'$\pi(a|x)$',fontsize=30,ha='center',va='center')
```
